```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

```python
total_revenue = df.groupby("Product")["Revenue"].sum()
print("Total Revenue per Product is: ")
display(total_revenue)
```

```python
avg_units = df.groupby("Product")["Units_Sold"].mean()
print("Average units sold per product.: ")
display(avg_units)
```

```python
revenue_trend = df.pivot(index="Month", columns= "Product", values= "Revenue")
month_order = ["Jan", "Feb", "Mar"]
revenue_trend.reindex(month_order)

plt.figure(figsize=(7,5))
for product in revenue_trend.columns:
    plt.plot(revenue_trend.index, revenue_trend[product], marker='o', label = product)


plt.title("Monthly Revenue Trend for each product (Jan–Mar)")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend(title = "Product")
plt.show()
```

```python
import matplotlib.pyplot as plt
total_revenue = df.groupby("Product")["Revenue"].sum()
plt.figure(figsize=(7,5))
total_revenue.plot(kind="bar", edgecolor = "black")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.title("Total revenue per product")
plt.show()
```

```python
plt.figure(figsize=(6,6))
total_revenue.plot(kind='pie', autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor':'black'})
plt.title("Revenue Contribution by Product")
plt.ylabel("")
plt.show()
```
